### 1.Write a Python Program to Find the Factorial of a Number?

```python
def fact(num):
    a=1
    while num>1:
        a*=num
        num-=1
    return a

fact(6)
```

### 4.Write a Python Program to Check Armstrong Number?

```python
def armstrong(num):
    temp1= str(num)
    temp2=0
    for i in temp1:
        temp2+=int(i)**3
    if temp2==num:
        print(f'{num} is an Armstrong number')
    else:
        print(f'{num} is not an Armstrong number')
```

```python
armstrong(153)
```
