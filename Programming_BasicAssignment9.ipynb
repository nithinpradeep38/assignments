1.	Write a Python program to check if the given number is a Disarium Number?

```python
def is_disarium(num):
    temp= str(num)
    count=1
    check=0
    for i in temp:
        check+= int(i)**count
        count+=1
    if num== check:
        return True
    return False
```

2.	Write a Python program to print all disarium numbers between 1 to 100?

```python
def print_disarium():
    for i in range(1,101):
           if is_disarium(i):
                print(i)
```

```python
print_disarium()
```

3.	Write a Python program to check if the given number is Happy Number?

```python
def happy_num(num):
    temp= str(num)
    square=0
    
    for i in temp:
        square+=int(i)**2
    
    if square==1:
        return True
    elif square>=10:
        return happy_num(square)
        #return square
        
    else:
        return False
```

```python
happy_num(19)
```

5.	Write a Python program to determine whether the given number is a Harshad Number?

```python
def harshad(num):
    sumdigit= 0
    for i in str(num):
        sumdigit+= int(i)
    if num%sumdigit==0:
        return True
    return False
```

```python
harshad(2020)
```

6.	Write a Python program to print all pronic numbers between 1 and 100?

```python
def print_pronic():
    num=1
    count=1
    while count:
        num= count*(count+1)
        count+=1
        if num<=100:
            print(num)
        else:
            break
```

```python
print_pronic()
```
