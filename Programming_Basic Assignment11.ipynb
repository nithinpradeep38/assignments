1.	Write a Python program to find words which are greater than given length k?

```python
def stringfind(list_of_words,k):
    temp= []
    for i in list_of_words:
        if len(i)>k:
            temp.append(i)
    return temp
```

```python
stringfind(['abc', 'abcde', 'abcdef', 'a', 'abcdefg'], 3)
```

2.	Write a Python program for removing i-th character from a string?

```python
def removechar(word,i):
    return word[:i-1]+word[i:]
```

3.	Write a Python program to split and join a string?

```python
def splitstring(string, i):
    return [string[:i], string[i+1:]]
```

```python
splitstring("ineuron",2)
```

```python
def joinstring(string1, string2):
    return string1+string2
```

```python
joinstring("fsds", "ineuron")
```

4.	Write a Python to check if a given string is binary string or not?

```python
def binarystring(string):
    for i in string:
        if i not in ['1', '0']:
            return False
    return True
```

```python
binarystring('10001')
```

```python
binarystring('101011b')
```

5.	Write a Python program to find uncommon words from two Strings?

```python
def uncommon(string1,string2):
    temp1= string1.split()
    temp2= string2.split()
    temp1.extend(temp2)
    
    unique=[]
    [unique.append(x) for x in temp1 if x not in unique]
    
    return unique
```

```python
uncommon("ineuron data science", "full stack data science")
```

6.	Write a Python to find all duplicate characters in string?

```python
def dupechar(string):
    duplicate=[]
    for i in range(len(string)-1):
        for j in range(i+1, len(string)):
            if string[i]== string[j]:
                duplicate.append(string[i])
    return set(duplicate)
```

```python
dupechar("ineuron")
```

```python
dupechar("nithin pradeep")
```

7.	Write a Python Program to check if a string contains any special character?

```python
def special_char(string):

    special_char= [list(range(32,48)),list(range(58,65)),list(range(91,97)),list(range(123,127)) ] #ASCII for special char
    special_char_unpacked= [x for i in special_char for x in i] #list of ASCII of special characters
    
    for i in string:
        if ord(i) in special_char_unpacked:
            return True
        
    return False

```

```python
special_char("Nithin")
```

```python
special_char("Hi! How are you?")
```
