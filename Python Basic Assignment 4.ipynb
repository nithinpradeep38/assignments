9. What are the list concatenation and list replication operators?

We can use + for concatenation of lists. If a and b are lists, a+b gives a list concatenation of a & b. * operator can be used for replication. 

10. What is difference between the list methods append() and insert()?

append() takes the element to be added as the argument and adds the element to the end of the list. Insert function requires two arguments- The index and the element to be inserted. It inserts the element at the position we ask it to be inserted in the list.

11. What are the two methods for removing items from a list?

pop()--> Removes the last element of the list.

remove()--> Takes the element to be removed as argument, checks if the element is there in the list and removes its first occurence. If the element is not available, it raises an error.

13.What's the difference between tuples and lists?

Tuple is a set of immutable collection of elements. Lists are mutable collection of elements.

15. How do you get a list value's tuple form? How do you get a tuple value's list form?

List value's tuple form can be obtained using tuple function.

Similarly, tuple value's list form can be obtained using list function.

list(tuple_name)

16.Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain?

They are references which bind the variable name to the list object.

17. How do you distinguish between copy.copy() and copy.deepcopy()?

copy.copy() only creates a new object with references to the original object.
copy.deepcopy() creates a new object and replicates the original object by creating copies.
