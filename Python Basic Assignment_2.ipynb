### 10. How can you tell the difference between break and continue?

A break statements terminates the execution of the loop, exits the loop and then control goes to the statement succeeding the loop.
A continue loop simply skips the current iteration when it is encountered. The loop is not exited. Control goes back to the loop with next iteration.
