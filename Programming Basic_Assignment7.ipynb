3.Write a Python Program for array rotation?

```python
#rotation to the left by n times


def arrayrotate(arr):
    temp= arr[0]
    for i in range(len(arr)-1):
        arr[i]= arr[i+1]
        
    arr[len(arr)-1]= temp
    
    
def multirotate(arr,n):
    for i in range(n):
        arrayrotate(arr)
        print(f"Rotated array after {i+1 } rotations is ", arr )
```

```python
l= [1,2,3,4,5]
multirotate(l,2)
```

4.	Write a Python Program to Split the array and add the first part to the end?

```python
def splitarray(l, pos):
    if pos>len(l):
        print("Cannot do the operation")
    else:
        temp1= l[:pos]
        temp2= l[pos:]
        temp2.extend(temp1)
        print("Final array is", temp2)
```

```python
l=[1,2,3,4,5]
splitarray(l,2)
```

5.Write a Python Program to check if given array is Monotonic?

```python
def monotonic(arr):
    flag1=1
    for i in range(len(arr)-1):
        if arr[i]> arr[i+1]:
            flag1=0
            break
    if flag1==1:
        print("Monotonically increasing")
        
    else:
        for j in range(len(arr)-1):
            if arr[j]< arr[j+1]:
                flag1=1
                break
        if (flag1==0):
            print("Monotonically decreasing")
        else:
            print("Non-monotonic")
        
    
        
```

```python
monotonic([4,5,9,10])
```

```python
monotonic([7,7,6,5])
```

```python
monotonic([3,2,7,4])
```
