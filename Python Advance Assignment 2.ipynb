### 1.What is the relationship between classes and modules?

The class is defined using the keyword class and can be instantiated by creating an instance.
A module is basically a chunk of python code which can have its own variables, functions etc., and imported using the keyword import. Unlike class, it cannot be instantiated.

### 3.Where and how should be class attributes created?

Class attributes are variables belonging to the class rather than the class instance. These are shared by all the instances.

Class attributes are typically defined at the top of the body of class definition( right below the class name.)

### 4. Where and how are instance attributes created?

These are attributes belonging inside the scope of a particular instance. They are defined at the time of creation of instance using the __init__ constructor.

### 5. What does the term "self" in a Python class mean?
 
self is a reference to a particular instance. We can access the class instance attributes using self.

### 6. How does a Python class handle operator overloading?

Python has special methods that are called depening upon the situation.  For example, the + operator is overloaded using a special method named __add__(). This method is invoked by both str and int classes.


### 7. When do you consider allowing operator overloading of your classes?

We use operator overloading of classes when we want to give special meaning to the operators. For example, we want to add two objects in a certain manner, we can overload the + operator to be given special meaning to carry addition in a user-defined way.

### 8. What is the most popular form of operator overloading?

A very popular form of operator overloading is that of + operator where it is used for addition between numbers as well as string concatenation.
