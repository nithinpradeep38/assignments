```python
#1.	Write a Python program to convert kilometers to miles?
def kmtomiles(km):
    try:
        return(0.621*km)
    except Exception as e:
        print(e)
        
        
kmtomiles(5)
```

```python
#2.	Write a Python program to convert Celsius to Fahrenheit?
def celstofar(cel):
    try:
        return((cel*(9/5))+32)
    except Exception as e:
        print(e)
        
        
celstofar(0)
```
